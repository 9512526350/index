```python
import matplotlib.pyplot as plt
```

```python
# Visa Applications Per Month (Bar Chart)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
applications = [120, 150, 180, 200, 170, 190]

plt.figure(figsize=(8,5))
plt.barh(months, applications, color=['#ff5733', '#33ff57', '#3357ff', '#ff33a8', '#ffdb33', '#8d33ff'])
plt.xlabel("Months")
plt.ylabel("Visa Applications")
plt.title("Visa Applications Per Month")
plt.show()
```

```python
# Visa Approval vs. Rejection Rate (Pie Chart)
labels = ["Approved", "Rejected"]
counts = [450, 50]
colors = ["#2ecc71", "#e74c3c"]

plt.figure(figsize=(6,6))
plt.pie(counts, labels=labels, autopct="%1.1f%%", colors=colors, startangle=140, wedgeprops={'edgecolor': 'white'})
plt.title("Visa Approval vs. Rejection Rate")
plt.show()
```

```python
# Visa Processing Time (Line Chart)
visa_types = ["Tourist", "Business", "Student", "Work"]
processing_days = [10, 15, 30, 45]

plt.figure(figsize=(8,5))
plt.plot(visa_types, processing_days, marker='o', linestyle='-', color='#9b59b6', markersize=8, linewidth=2)
plt.xlabel("Visa Type")
plt.ylabel("Processing Time (Days)")
plt.title("Average Visa Processing Time by Type")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Sales Trends Over 6 Months (Line Chart)
sales = [20000, 25000, 30000, 28000, 35000, 40000]

plt.figure(figsize=(8,5))
plt.plot(months, sales, marker='o', linestyle='-', color='#f39c12', markersize=8, linewidth=2)
plt.xlabel("Months")
plt.ylabel("Sales Revenue ($)")
plt.title("E-Commerce Sales Trends Over 6 Months")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Top 5 Best-Selling Products (Horizontal Bar Chart)
products = ["Laptop", "Phone", "Shoes", "Watch", "Bag"]
units_sold = [150, 200, 250, 180, 220]

plt.figure(figsize=(8,5))
plt.barh(products, units_sold, color=['#3498db', '#e67e22', '#1abc9c', '#9b59b6', '#f1c40f'])
plt.xlabel("Units Sold")
plt.ylabel("Products")
plt.title("Top 5 Best-Selling Products")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Orders by Category (Pie Chart)
categories = ["Electronics", "Clothing", "Home Appliances", "Books", "Accessories"]
orders = [300, 500, 200, 150, 250]
colors = ["#e84118", "#44bd32", "#00a8ff", "#8c7ae6", "#fbc531"]

plt.figure(figsize=(6,6))
plt.pie(orders, labels=categories, autopct="%1.1f%%", colors=colors, startangle=140, wedgeprops={'edgecolor': 'white'})
plt.title("Orders by Category")
plt.show()
```
